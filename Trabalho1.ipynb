# Método para plotar os gráficos

```
def plotGrafico(f, inicio, final, raizes, metodo):
  
  numero_pontos = 100

  curve = np.linspace(inicio,final,numero_pontos+1)
  
  values = [f(value) for value in curve]
  
  x_aproximado = [f(value) for value in raizes]
  
  
  plt.plot(curve, values, linestyle='--', color = 'blue', alpha = 0.7)

  plt.plot(raizes, x_aproximado, 'x', marker='o', label='' + str(len(raizes) - 1) + ' iterações', color = 'green' )

  i=0
  for x,y in zip(raizes, x_aproximado):
    i = i+1
    plt.text(x,y,str(i))

  leg = plt.legend();
  
  plt.axhline(0, color = 'gray')
  plt.axvline(0, color = 'gray')
  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title('Método da ' + metodo)

  plt.axhline(0, color = "black")

  plt.grid()

  plt.show()
```

# Método da Bisseção

```
def bissecao (f, a, b, precisao):
  n = 0
  raizes= []
  inicio = a
  final = b
  maxIteracao = 500
  
  if (f(a) * f(b)) < 0:
    x = (a + b)/2
    
    print("%-5s%-10s%-9s%-8s%-10s%-30s"%("n","a", "b", "x", "f(x)", "f(a)"))
    raizes.append(x) 
    while abs(f(x)) > precisao:
      if f(a) * f(x) < 0:
        b = x
      elif f(b) * f(x) < 0: 
        a = x
      n += 1  
      x = (a + b) / 2
      raizes.append(x)    
     
      print("%d %f %f %f %f %f" %(n, a, b, x, f(x), f(a)))
      
      if(n >= maxIteracao):
        break
 
    print("A raiz do intervalo é", x)
    
  else:
    print("Não contém raizes no intervalo dado") 
    
  plotGrafico(f, inicio, final, raizes, "Bisseção")
```

```
#Questão 2a
f1 = lambda x: x**3 - 6*x**2 - x + 30 
bissecao(f1, 2, 4, 1e-2)

#Questão 2b
f2 = lambda x: x + math.log(x)
bissecao(f2, 0.3, 0.5, 1e-2)

#Questao 2c
f3 = lambda x: 3*x - math.cos(x)
bissecao(f3, -2, 2, 1e-2)

#Questão 2d
f4 = lambda x: x + 2*math.cos(x)
bissecao(f4, -5, 0, 1e-2)

#Questão 2e
f5 = lambda x: x**3 - 5*x**2 + 17*x + 21
bissecao(f5, -1, 0, 1e-2)

#Questão 2f
f6 = lambda x: x**2 - 3
bissecao(f6, 1, 2, 1e-2)

#Questão 2g
f7 = lambda x: x**2 + math.log(x)
bissecao(f7, 0.5, 1, 1e-2)

#Questão 2h
f8 = lambda x: x**3 -10
bissecao(f8, 2, 3, 1e-2)

#Questão 2i
f9 = lambda x: 4*math.sin(x) - math.exp(x)
bissecao(f9, -5, 5, 1e-3)
```

# Método Falsa Posição

```
def fp (f, a, b, precisao):
  n = 0
  inicio = a
  final = b
  raizes = []
  maxIteracao = 500
  
  if (f(a) * f(b)) < 0:
    x = (a*f(a) - b*f(b))/(f(a) - f(b))
    raizes.append(x) 
    print("%-5s%-10s%-9s%-8s%-10s%-30s"%("n","a", "b", "x", "f(x)", "f(a)"))
    while abs(f(x)) > precisao:
      if f(a) * f(x) < 0:
        b = x
      elif f(b) * f(x) < 0: 
        a = x
      x = (a*f(a) - b*f(b))/(f(a) - f(b))
      n += 1
      raizes.append(x)
      print("%d %f %f %f %f %f" %(n, a, b, x, f(x), f(a))) 
      
      if(n >= maxIteracao):
        break
            
    print("A raiz do intervalo é", x)
  else:
    print("Não contém raizes no intervalo dado")
    
  plotGrafico(f, inicio, final, raizes, "Falsa Posição")
    
  
```

```
#Questão 1a - Lista 3
f1 = lambda x: math.exp(math.cos(x)) + x**3 - 3
fp(f1, 1.1, 1.2, 1e-3)

#Questo 1b - Lista 3
f2 = lambda x: 0.1*x**3 - math.exp(2*x) + 2
fp(f2, 8, 10, 1e-3)

#Questão 1c - Lista 3
f3 = lambda x: 2*math.log(3 - math.cos(x)) - 3*x + 5*math.sin(x)
fp(f3, 2, 3, 1e-3)

#Questão 1d - Lista 3
f4 = lambda x: x**3 - 5*x**2 + x + 3
fp(f4, 0, 2, 1e-3)

```

# Método das Secantes

```
def secantes(f, x0, x1, precisao):
  raizes = []
  inicio = x0
  final = x1
  maxIteracao = 500
  k = 0
  
  if(abs(f(x0)) < precisao):
    print("A raiz é", x0)
  
  elif(abs(f(x1)) < precisao):
    print("A raiz é", x1)
  
  print("k\t x\t f(x)")
  print("%d %f %f" %(k, x0, f(x0)))
  k += 1
  print("%d %f %f" %(k, x1, f(x1)))
  
  raizes.append(x0)
  raizes.append(x1)
    
  x2 = x1 - ((f(x1)/(f(x1) - f(x0))) * (x1 - x0))
  raizes.append(x2)
  k += 1
  print("%d %f %f" %(k, x2, f(x2)))
  
  
  while abs(f(x2)) > precisao:
    x0 = x1
    x1 = x2   
    x2 = x1 - ((f(x1)/(f(x1) - f(x0))) * (x1 - x0))
    raizes.append(x2)
    k += 1
    print("%d %f %f" %(k, x2, f(x2)))
    
    if(k >= maxIteracao):
      break
  
  print("A raiz do intervalo é", x2)
  plotGrafico(f, inicio, final, raizes, "Secantes")
```

```
# Questão 2a - Lista 3
f1 = lambda x: x**2 - 10*math.log(x) - 5
secantes(f1, 4, 5, 1e-3)

# Questão 2b - Lista 3
f2 = lambda x: x**3 - math.exp(2*x) + 3
secantes(f2, 0.5, 1, 1e-3)

# Questão 2c - Lista 3
f3 = lambda x: 2*x**3 + x**2 - 2
secantes(f3, 0.8, 1, 1e-3)

# Questão 2d - Lista 3
f4 = lambda x: math.sin(x) - math.log(x)
secantes(f4, 2, 3, 1e-3)
```

# Método da Tangente

```
def tangente(f, fd, x, x_1, precisao):
  raizes = []
  inicio = x
  final = x_1
  maxIteracao = 500
  k = 0
  if abs(f(x)) < precisao:
    print("A raiz é ", x) 
    raizes.append(x)
  print("k\t x0\t f(x0)")
  while abs(f(x)) > precisao: 
    x = x - (f(x).evalf()/fd(x).evalf()) 
    raizes.append(x)
    print("%d %f %f" %(k, x, f(x)))
    if abs(f(x)) < precisao:
      print("A raiz do intervalo é", x)
    
    k += 1 
    if(k >= maxIteracao):
      break
  
  plotGrafico(f, inicio, final, raizes, "Tangente")
  
```

- Nos métodos da bisseção e da falsa posição podemos analisar as maiores complexidades de cálculos, primeiramente no método da bisseção vai ter uma dependência maior numa média das raízes, ou seja, vai depender de cálculos de médias passadas dos intervalos para chegar na raiz mais aproximada.

- No método da falsa posição vai ocorrer uma maior iteração por conta da complexidade dada nos cálculos nas raizes, nesse caso os cálculos vão depender de funções de F(a) e F(b) para encontrar a raiz mais aproximada.

- Algumas das funções, como a função de número 4, podem nos gerar uma grande iteração entre os valores obtidos das raizes para ser calculada a mais aproximada.

- Os métodos apresentados são definidos a partir de intervalos das funções para o cálculo das raízes, nisso para um dado erro(precisão) das raízes os métodos vão sendo calculado a partir de uma recursão da f(x) > precisao onde irá para quando chegar na f(x) mais aproximada do eixo x tendo assim a raiz desejada.
